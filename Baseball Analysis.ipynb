### An analysis of the predictive value of spring training statisitics on regular season outcomes.

###The Question
####Are Spring Training Statistics predictive of regular season results?

Every year in North America, the month of April carries within it the inauguration of a new season of Major League Baseball. But before the first pitch is thrown and the 162 game season gets underway, a whole other "practice" season is played, known as spring training. 

Spring Training runs from mid February to late March and consists of between 30 to 40 games across 2 separate leagues based in the American Southwest (Cactus League) and the Southeast (Grapefuit League). The 2 spring leagues are comprised of a mix of teams from both the regular season National and American League divisions. There are many potential problems with attempting to utilize spring training stats as a predictor of how successful a player or team will perform during the regular season. Everything from a small sample (only 30 or so games), to players experimenting with batting and pitching approaches, to questions of how competitive the games really are when winning holds no concrete value. However, the idea that there is some predictive value to spring training statisitics is one that is always fun to entertain.

###The Data
Logistic regression -- probabilities of binomial events.

Since our response variable (PostSeason) is a binary variable I'll use logistic regression to predict whether a team makes the postseason playoffs 

Logistic regression plot with PostSeason as response variable and spring training winning percentage as predictor

```python
sns.lmplot("WPct", "PostSeason", bb, y_jitter=.05, logistic=True);
```

```python
logm = linear_model.LogisticRegression()

X = bb[['WPct',	'R', 'H', 'x2B', 'x3B', 'HR', 'RBI', 'AVG', 'OBP', 'SLG', 'OPS']].values
y = bb.PostSeason
```

```python
logm.fit(X, y)
```

```python
logm.predict(X)
```

```python
logm.score(X, y)
```

So, from the looks of our score, spring training stats only slightly improve the probability of a team making the post season.
