###The Question
####Are Spring Training Statistics predictive of regular season results?

Every year in North America, the month of April carries within it the inauguration of a new season of Major League Baseball. But before the first pitch is thrown and the 162 game season gets underway, a whole other season is played as a warm-up known as Spring Training. 

Spring Training runs from mid February to late March and consists of between 30 to 40 games across 2 separate leagues based in the American Southwest (Cactus League) and the Southeast (Grapefuit League). The 2 spring leagues are comprised of a mix of teams from both the regular season National and American League divisions. There are many potential problems with attempting to utilize Spring Training stats as a predictor of how successful a player or team will perform during the regular season. Everything from a small sample (only 30 or so games), to players experimenting with batting and pitching approaches, to questions of how competitive the games really are when winning holds no concrete value. However, the idea that there is some predictive value to Spring Training statisitics is one that should be fun to entertain.
